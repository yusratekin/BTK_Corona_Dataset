# YARIŞAN GRAFİKLER

```python
#Veri Yükle
df=pd.read_csv("corona_dat.csv")
```

```python
df['date'].dtype
```

```python
#Date sütununun tarih veri tipine çevirme
import datetime
df['date']=pd.to_datetime(df['date'])
df['date'].dtype
```

```python
# Veri türlerini kontrol etme
df.info()
```

```python
df_yarisan_grf=df[['date','Turkey','Italy','Spain','Germany','China','Russia','US']]
```

```python
df_yarisan_grf.head()
```

```python
# İndex Tarih sırasına göre girilmiş ama her zmaan bu şekilde olmayabilir
# O yüzden date tatrhi formatına çevrilir ve tarihe göre sıralama (index) yapılır
df_yarisan_grf.set_index('date',inplace=True)
df_yarisan_grf.head() # index numarası yerine tarih kullanıldı
```

```python
import seaborn as sns
#Günlük vaka sayısı
sns.lineplot(data=df_yarisan_grf);
```

```python
df_yarisan_grf=df_yarisan_grf.cumsum(axis=0)
df_yarisan_grf
```

```python
sns.lineplot(data=df_yarisan_grf,marker='*');
plt.xticks(rotation=60)
plt.title("Günlük vaka sayısı");
plt.show()
```
